```python
# Entraîner et évaluer les modèles
models = {'dummy_regr_mean': DummyReg_mean,
          'dummy_regr_median': DummyReg_median,
          'linearr': LinearR,
          'lasso': Lasso,
          'ridge': Ridge,
          'elastic': Elastic,
          'GB': GardienBoostingR,
          'rfr': RFR,
          'ETR': ExtraTR}

results = {}
for name, model in models.items():
    # Début de la mesure du temps
    start_time = time.time()
    model.fit(X_train, y_train)
     # Fin de la mesure du temps
    end_time = time.time()
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_test_pred)
    tr_mae = median_absolute_error(y_train, y_train_pred)
    valid_mae = median_absolute_error(y_test, y_test_pred)
    train_rmse = mean_squared_error(y_train,y_train_pred)**0.5
    valid_rmse = mean_squared_error(y_test,y_test_pred)**0.5
    # Calcul du temps d'entraînement
    training_time = end_time - start_time
    results[name] = [round(r2,3),round(tr_mae,3), round(valid_mae,3), round(train_rmse,3), round(valid_rmse,3),training_time]
    evaluation(model)
```

* Mean Absolute Error (MAE) : MAE signifie erreur absolue moyenne. Il s'agit d'une mesure de performance couramment utilisée pour les modèles de régression pour évaluer la différence entre les valeurs prévues et les valeurs réelles. Il est calculé en prenant la somme des erreurs absolues pour chaque observation et en divisant par le nombre total d'observations.
* RMSE signifie erreur quadratique moyenne. Il s'agit d'une autre mesure de performance couramment utilisée pour les modèles de régression pour évaluer la différence entre les valeurs prévues et les valeurs réelles. Il est calculé en prenant la somme des erreurs au carré pour chaque observation, en divisant par le nombre total d'observations et en prenant la racine carrée du résultat.

L'erreur quadratique moyenne (RMSE) est une mesure de l'adéquation entre les valeurs prédites d'un modèle et les valeurs réelles des données. Un score RMSE plus faible indique une meilleure adéquation entre les valeurs prédites et les valeurs réelles.

Un score de 0 pour le RMSE sur l'ensemble d'apprentissage signifierait que le modèle est parfaitement adapté aux données d'apprentissage, ce qui est peu probable en pratique. Il peut s'agir d'une indication de sur apprentissage, ce qui signifie que le modèle a appris le bruit des données d'apprentissage et qu'il ne se généralise pas bien aux nouvelles données.

Il est important de garder à l'esprit qu'un bon modèle ne doit pas seulement avoir un faible score RMSE sur l'ensemble d'apprentissage mais aussi sur l'ensemble de validation ou de test. L'objectif d'un modèle est de bien généraliser à de nouvelles données non vues, et non de s'adapter parfaitement à l'ensemble d'apprentissage.

Il est également important de vérifier d'autres mesures d'évaluation telles que le R-squared, l'erreur absolue moyenne ou l'erreur quadratique moyenne.

Vous devez également vérifier le tracé des résidus si possible, c'est une bonne pratique de vérifier les résidus pour voir s'ils sont dispersés de façon aléatoire autour de zéro ou s'il y a un modèle dans les résidus qui pourrait indiquer une mauvaise spécification du modèle.

# Model vs ENERGYSTARScore 

Le score ENERGY STAR est un système de notation développé par l'Agence américaine pour la protection de l'environnement (EPA) pour évaluer l'efficacité énergétique des bâtiments commerciaux. Les bâtiments peuvent obtenir un score compris entre 1 et 100, un score de 75 ou plus indiquant que le bâtiment est plus performant qu'au moins 75 % des bâtiments similaires dans le pays. Le score est basé sur l'intensité de l'utilisation d'énergie (EUI) du bâtiment, qui est calculée en divisant la consommation totale d'énergie du bâtiment par sa superficie brute en pieds carrés.

```python
# Entraîner et évaluer les modèles
models = {'linearr': LinearR,
          'ridge': ridge,
          'GB': GardienBoostingR,
          'rfr': RFR,
          'ETR': ExtraTR}


results_energyscore = {}
for name, model in models.items():
    # Début de la mesure du temps
    start_time = time.time()
    model.fit(X_train, y_train)
     # Fin de la mesure du temps
    end_time = time.time()
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    r2 = r2_score(y_test, y_test_pred)
    tr_mae = median_absolute_error(y_train, y_train_pred)
    valid_mae = median_absolute_error(y_test, y_test_pred)
    train_rmse = mean_squared_error(y_train,y_train_pred)**0.5
    valid_rmse = mean_squared_error(y_test,y_test_pred)**0.5
    # Calcul du temps d'entraînement
    training_time = end_time - start_time
    results_energyscore[name] = [round(r2,3),round(tr_mae,3), round(valid_mae,3), round(train_rmse,3), round(valid_rmse,3),training_time]
    evaluation_energyscore(model)
```

```python
# Stocker les résultats dans un dataframe
results_energyscore_df = pd.DataFrame(results_energyscore).T
results_energyscore_df.columns = ["R2","Train MAE", "Validation MAE","Train RMSE", "Valid RMSE",'Temps']
```
